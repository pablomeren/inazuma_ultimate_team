<h1>Pretendemos crear un equipo que contenga los 11 jugadores con mayores estadísticas respetando su posición y utilizando la formación: 1-4-3-3</h1>

```python
positions = ie['Position'].unique()
positions
```

```python
stats_cols = ['Kick', 'Body', 'Control', 'Guard', 'Speed', 'Stamina', 'Guts']
ie['overall'] = ie[stats_cols].mean(axis=1)
def mejores_por_posicion(ie, posicion, n):
     return (
        ie[ie['Position'] == posicion]
        .sort_values('overall', ascending=False)
        .head(n)
    )
```

```python
# Mejor portero
mejor_gk = mejores_por_posicion(ie, 'GK', 1)
mejor_gk
```

```python
#Mejores 4 defensas
mejor_df = mejores_por_posicion(ie, 'DF', 4)
mejor_df
```

```python
#Mejores 3 centrocampistas
mejor_mf = mejores_por_posicion(ie, 'MF', 3)
mejor_mf
```

```python
#Mejores 3 delanteros
mejor_fw  = mejores_por_posicion(ie, 'FW', 3)
mejor_fw
```

```python
equipo_definitivo = pd.concat([mejor_gk, mejor_df, mejor_mf, mejor_fw])
equipo_definitivo
```
